```python
import pandas as pd
```

```python
df = pd.read_csv('Loan_default.csv', index_col=0)
pd.set_option("display.max_columns",None)

df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
categorical_features=['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage', 'HasDependents', 'LoanPurpose','NumCreditLines', 'HasCoSigner', 'Target']

numerical_features=['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'InterestRate', 'LoanTerm', 'DTIRatio']
```

```python
for variable in categorical_features:
    print(df[variable].value_counts())
    print("~"*50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20, 20))
ind=1
for variable in categorical_features:
    plt.subplot(3,3,ind)
    sns.countplot(x=df[variable],palette='Set1')
    plt.ylabel('Count')
    plt.xlabel(variable)
    ind+=1
plt.tight_layout()
plt.show()
```

### Категориальные представлены в одинаковом количестве
### Дисбаланс классов

```python
plt.figure(figsize=(20, 20))
ind=1
for variable in categorical_features[:8]:
    plt.subplot(3,3,ind)
    sns.countplot(x=df[variable],palette='Set2',hue=df["Target"])
    plt.ylabel('Count')
    plt.xlabel(variable)
    ind+=1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 20))
ind=1
for variable in numerical_features:
    plt.subplot(3, 3, ind)
    sns.histplot(x=df[variable],kde=True,color="green")
    plt.title(variable)
    ind+=1
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df[['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'InterestRate', 'LoanTerm','NumCreditLines' ,'DTIRatio',"Target"]].corr(), annot=True, cmap='Reds')
plt.title('Correlation Matrix')
plt.show()
```
